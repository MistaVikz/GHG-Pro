```python
# Generate random default rate/ recovery potential tables
import random
import pandas as pd

def generate_default_rates_and_recovery_potential():
    """
    This function generates two dictionaries: default_rates and recovery_potential.
    
    The default_rates dictionary has increasing float values while the integer keys increase.
    The "Investment" values are less than the "Speculative" values, which are less than the "C" values.
    
    The recovery_potential dictionary has decreasing float values while the integer keys increase.
    The "C" values are less than the "Speculative" values, which are less than the "Investment" values.
    
    All float values are between 0 and 1.
    
    Returns:
        default_rates (dict): A dictionary of default rates for investments, speculative, and C ratings.
        recovery_potential (dict): A dictionary of recovery potential for investments, speculative, and C ratings.
    """

    default_rates = {
        "Investment": {},
        "Speculative": {},
        "C": {}
    }

    recovery_potential = {
        "Investment": {},
        "Speculative": {},
        "C": {}
    }

    for i in range(1, 11):
        if i == 1:
            investment_rate = round(random.uniform(0.01, 0.10), 2)
            speculative_rate = round(random.uniform(investment_rate + 0.01, 0.20), 2)
            c_rate = round(random.uniform(speculative_rate + 0.01, 0.30), 2)

            investment_recovery = round(random.uniform(0.90, 0.99), 2)
            speculative_recovery = round(random.uniform(0.80, investment_recovery - 0.01), 2)
            c_recovery = round(random.uniform(0.70, speculative_recovery - 0.01), 2)
        else:
            investment_rate = round(random.uniform(default_rates["Investment"][i-1] + 0.01, min(0.99, default_rates["Investment"][i-1] + 0.05)), 2)
            speculative_rate = round(random.uniform(max(investment_rate + 0.01, default_rates["Speculative"][i-1] + 0.01), min(0.99, default_rates["Speculative"][i-1] + 0.05)), 2)
            c_rate = round(random.uniform(max(speculative_rate + 0.01, default_rates["C"][i-1] + 0.01), min(0.99, default_rates["C"][i-1] + 0.05)), 2)

            investment_recovery = round(random.uniform(max(0.01, recovery_potential["Investment"][i-1] - 0.05), recovery_potential["Investment"][i-1] - 0.01), 2)
            speculative_recovery = round(random.uniform(max(0.01, recovery_potential["Speculative"][i-1] - 0.05), recovery_potential["Speculative"][i-1] - 0.01), 2)
            c_recovery = round(random.uniform(max(0.01, recovery_potential["C"][i-1] - 0.05), recovery_potential["C"][i-1] - 0.01), 2)

        default_rates["Investment"][i] = investment_rate
        default_rates["Speculative"][i] = speculative_rate
        default_rates["C"][i] = c_rate

        recovery_potential["Investment"][i] = investment_recovery
        recovery_potential["Speculative"][i] = speculative_recovery
        recovery_potential["C"][i] = c_recovery

    return default_rates, recovery_potential

default_rates, recovery_potential = generate_default_rates_and_recovery_potential()

# Convert dictionaries to DataFrames
df_default_rates = pd.DataFrame(default_rates).T
df_recovery_potential = pd.DataFrame(recovery_potential).T

# Save DataFrames to CSV files
df_default_rates.to_csv('default_rates.csv')
df_recovery_potential.to_csv('recovery_potential.csv')
```
